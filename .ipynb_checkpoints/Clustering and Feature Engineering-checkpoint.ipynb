# VideoStarted Data Feature Enginnering

```python
VideoStarted['Date'] = pd.to_datetime(VideoStarted['Date'],format='%Y%m%d')

'''First Video viewed Date '''
Video_start = pd.DataFrame(VideoStarted.groupby('UserId')['Date'].min())
Video_start.rename(columns={'Date':'Start_Date'},inplace=True)


'''Last Video viewed Date'''
Video_End = pd.DataFrame(VideoStarted.groupby('UserId')['Date'].max())
Video_End.rename(columns={'Date':'Last_Date'},inplace=True)


'''Creating User wise data from Video Started Data'''
Video_Final = pd.merge(Video_start,Video_End,left_on='UserId',right_on='UserId',how='left')
Video_Final['User_Duration'] = ((Video_Final['Last_Date'] - Video_Final['Start_Date']) / np.timedelta64(1,'h'))/24

'''Unique videos seen by user'''
Video_unique = pd.DataFrame(VideoStarted.groupby('UserId')['VideoId'].nunique())
Video_unique.rename(columns={'VideoId':'Unique_Videos_Watched'},inplace=True)

'''Merging Unique videos to Final Video DataFrame'''
Video_Final = pd.merge(Video_Final,Video_unique,left_on='UserId',right_on='UserId',how='left')

'''No of videos watched by User'''
Videos_Watched = pd.DataFrame(VideoStarted['UserId'].value_counts())
Videos_Watched.rename(columns={'UserId':'Total_Videos_Watched'},inplace=True)
Videos_Watched.reset_index(inplace=True)
Videos_Watched.rename(columns={'index':'UserId'},inplace=True)

'''Final Video Data'''
Video_Final = pd.merge(Video_Final,Videos_Watched,left_on='UserId',right_on='UserId',how='left')
Video_Final.head()
```

# AppLaunched Data Feature Engineering

```python
from datetime import datetime
startTime = datetime.now()

'''App Launch Count'''
AppLaunched_Count = pd.DataFrame(AppLaunched['UserId'].value_counts())
AppLaunched_Count.rename(columns={'UserId':'launch_Count'},inplace=True)
AppLaunched_Count.index.rename('UserId',inplace=True)

'''Converting to Datetime Format'''
AppLaunched['Date'] = pd.to_datetime(AppLaunched['Date'],format='%Y%m%d')

'''First Launch'''
AppLaunched_Start = pd.DataFrame(AppLaunched.groupby('UserId')['Date'].min())
AppLaunched_Start.rename(columns={'Date':'AppFirstLaunch'},inplace=True)

'''Last Launch'''
AppLaunched_End = pd.DataFrame(AppLaunched.groupby('UserId')['Date'].max())
AppLaunched_End.rename(columns={'Date':'ApplastLaunch'},inplace=True)

'''Merging First and last launch DataFrame '''
AppLaunched_Final = pd.merge(AppLaunched_Start,AppLaunched_End,
                             left_on='UserId',right_on='UserId',how='left')
AppLaunched_Final['AppTimeSpan'] = ((AppLaunched_Final['ApplastLaunch'] - AppLaunched_Final['AppFirstLaunch']) /
                                      np.timedelta64(1,'h'))/24
AppLaunched_Final = pd.merge(AppLaunched_Final,AppLaunched_Count,
                             left_on='UserId',right_on='UserId',how='left')
'''Device Feature '''
AppOnDevice = AppLaunched.groupby('UserId')['Device'].value_counts().unstack()
AppOnDevice.index.rename('UserId',inplace=True)

AppLaunched_Final = pd.merge(AppLaunched_Final,AppOnDevice,
                             left_on='UserId',right_on='UserId',how='left')

'''Number of devices used by User'''
AppLaunched_Final['DevicesUsed'] = AppLaunched_Final[[0,1,2,3]].count(axis=1)

AppLaunched_Final.reset_index(inplace=True)

print('Time taken to run - {}'.format(datetime.now() - startTime))
```

```python
import datetime
str(datetime.timedelta(seconds=840))
AppLaunched['Time_Seconds'] = AppLaunched['Minute_Of_Day'] * 60 + AppLaunched['Second'] 
AppLaunched['Time'] = AppLaunched['Time_Seconds'].apply(lambda x : datetime.timedelta(seconds=x))
AppLaunched['DateTime'] = AppLaunched['Date'] + AppLaunched['Time']
```

```python
'''Vist duration Calculation'''
Visits_1 = pd.DataFrame(AppLaunched.groupby('UserId')['DateTime'].value_counts())
Visits_1.rename(columns={'DateTime':'Count'},inplace=True)
Visits_1.reset_index(inplace=True)
Visits_1.sort_values(by=['UserId','DateTime'],inplace=True)
Visits_1['Visit_Duration'] = round(Visits_1['DateTime'].diff()/np.timedelta64(1,'h'),0)

'''Making First Visit value of every user 0'''
bool_list = [True]
for i in range(len(Visits_1)):
    bool_list.append(Visits_1.iloc[i,0] == Visits_1.iloc[i+1,0])

bool_list = bool_list[0:-1]
```

```python
Visits_1.index = np.arange(0,len(Visits_1))
series = pd.DataFrame(bool_list)
mask = series[series[0]==False].index.tolist()
for i in mask:
    Visits_1.at[i,'Visit_Duration'] = 0

Visits_1['Visit_Duration_Final'] = Visits_1['Visit_Duration'].apply(lambda x : 0 if x<0 else x)
AppLaunched_Visit = pd.DataFrame(round(Visits_1.groupby('UserId')['Visit_Duration_Final'].mean(),0)).reset_index()
AppLaunched_Final = pd.merge(AppLaunched_Final,AppLaunched_Visit,
                             left_on='UserId',right_on='UserId',how='left')
AppLaunched_Final.rename(columns={'launch_Count':'Visit_Count','Visit_Duration_Final':'Avg_Time_for_Next_Visit',
                                 0:'AppDevice_Desktop',1:'AppDevice_Mobile',2:'AppDevice_Tablet',
                                 3:'AppDevice_TV'},inplace=True)
```

```python
from datetime import date
from datetime import time
from datetime import datetime

from functools import reduce
pd.set_option("display.max_rows", 500)

def time_stamp (df, date_col, min_col, sec_col, event_name):
    '''
    Returns dataframe with timestamp after combining the date, minute and sec column for the given dataframe
    '''
    df[date_col] = pd.to_datetime(df[date_col].astype(str))
    df[min_col] = pd.to_timedelta(df[min_col], unit='m')
    df[sec_col] = pd.to_timedelta(df[sec_col], unit='s')
    df[event_name] = df[date_col] + df[min_col] + df[sec_col]
    df.drop(columns=[date_col, min_col, sec_col], inplace=True)
    return df
```

## Cluster Analysis

```python
Video_Classification_1 = Video_Final[['User_Duration','Unique_Videos_Watched','Total_Videos_Watched','Genre Seen','Avg_Time_for_Next_Visit']]
Video_Classification_1.fillna(0,inplace=True)
from sklearn import preprocessing
x = Video_Classification_1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_test = pd.DataFrame(x_scaled)
df_test.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_test)
labels = kmeans.labels_

#Glue back to originaal data
df_test['clusters'] = labels
Video_Final['clusters'] = labels
```
